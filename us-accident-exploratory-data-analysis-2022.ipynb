###Analysis of Column Start Time

```python
df.Start_Time
```

```python
#Overwrite column to a datetime format
df.Start_Time =  pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
#Extracting hour of the day from times
start_time_accidents = df.Start_Time.dt.hour
start_time_accidents
```

```python
#Plotting a histplot for number of accidents happnedd in the given time of the day
sns.histplot(start_time_accidents, bins = 24)
```

```python
#Consider the percentage of accidents by hour
sns.distplot(start_time_accidents, bins = 24, kde = False, norm_hist=True)
```

A High number of accidents happen during 2-5 pm
The nwxt higher percentages are between 6-9 am

```python
#Consider the percentage of accidents by the day of the week(0- Monday)
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7, kde = False, norm_hist=True)
```

```python
#hourwise percentages on sundays
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek==6 ]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
#hourwise percentages on mondays
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek== 0]
sns.distplot(mondays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Q. Is the distribution of the hours same on weekends as on weekdays
on sunday peek occurs between 10am - 3pm unlike weekdays

```python
#Consider the percentages of accidents by months of the year
sns.distplot(df.Start_Time.dt.month, bins = 12, kde = False, norm_hist=True)
```

Can you explain the monthwise of trend of accidents
much data is missing for 2016 and even 2017

```python
#Consider the percentages of accidents by months of the year 2019
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist=True)
```
